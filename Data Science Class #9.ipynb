We can do much more interesting things with this approach. Let us add one more condition: that the tweet talks about Trump AND something of relevance to Palestine. 

```python
keywords1 =[" trump ", " donald "]
keywords2 =["palestin", "jerusalem", "israel"] 
```

Why do I write "palestin" and not "palestine"? Because we want all inflections of the word palestine (palestinians, palestinian, and so on). What about " trump " instead of "trump"? Because I don't want inflections to the word trump, like trumps (which is related to another verb, to trump). 

Let us go back to our example:
