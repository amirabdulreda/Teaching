Last week we looked at <b>for</b> and <b>if</b> arguments--today we will look at these arguments complemented with <b>else </b>and <b>elif</b> arguments. When we take them all together we can really start to do some serious data cleaning and analysis. 

Remember the logic of  the <b> for </b> statement. We use it in the following manner: 


```python
for <var> in <sequence>: 
    <body>
```

Remember also the logic of the <b> for </b> and <b> if </b> statements. 


```python
for <var> in <sequence>:
    if <cond>:
        <body>
```

With <b>else </b>and <b>elif</b> we will add some more conditions. We use else when we want the algorithm to finish with a given condition and elif when we want to add more than one condition. For instance: 

```python
for <var> in <sequence>:
    if <cond>:
        <body>
    else:
        <body> 
```

And for an elif loop:

```python
for <var> in <sequence>:
    if <cond>:
        <body>
    elif <cond>:
         <body>
    else: 
        <body>
```

Here is an example

```python
words_in_list =['Donald', 'Trump', 'lost', 'the', 'elections']
for x in words_in_list:
    if len(x)>3:
        print(x)
    else: 
        pass
```

Let's see how the pandemic affected changes in movements to and from retail & recreation in Turkey. We first have to create a mean value of each change in mobility by region.

```python
goglomore['retail_and_recreation_percent_change_from_baseline_reg'] = goglomore.groupby(['date', 'region'], as_index=False)['retail_and_recreation_percent_change_from_baseline_prov'].transform('mean')
goglomore['grocery_and_pharmacy_percent_change_from_baseline_reg'] = goglomore.groupby(['date', 'region'], as_index=False)['grocery_and_pharmacy_percent_change_from_baseline_prov'].transform('mean')
goglomore['parks_percent_change_from_baseline_reg'] = goglomore.groupby(['date', 'region'], as_index=False)['parks_percent_change_from_baseline_prov'].transform('mean')
goglomore['transit_stations_percent_change_from_baseline_reg'] = goglomore.groupby(['date', 'region'], as_index=False)['transit_stations_percent_change_from_baseline_prov'].transform('mean')
goglomore['workplaces_percent_change_from_baseline_reg'] = goglomore.groupby(['date', 'region'], as_index=False)['workplaces_percent_change_from_baseline_prov'].transform('mean')
goglomore['residential_percent_change_from_baseline_reg'] = goglomore.groupby(['date', 'region'], as_index=False)['residential_percent_change_from_baseline_prov'].transform('mean')
```

We then remove the averages by province so we can focus on the averages by regions. 

```python
del goglomore['province'] #Removing provinces so we can focus on regions for averages
del goglomore['retail_and_recreation_percent_change_from_baseline_prov'] #Removing provinces so we can focus on regions for averages
del goglomore['grocery_and_pharmacy_percent_change_from_baseline_prov'] #Removing provinces so we can focus on regions for averages
del goglomore['parks_percent_change_from_baseline_prov'] #Removing provinces so we can focus on regions for averages
del goglomore['transit_stations_percent_change_from_baseline_prov'] #Removing provinces so we can focus on regions for averages
del goglomore['workplaces_percent_change_from_baseline_prov'] #Removing provinces so we can focus on regions for averages
del goglomore['residential_percent_change_from_baseline_prov'] #Removing provinces so we can focus on regions for averages
```

Let us start by plotting the average change in mobility toward retail & recreation at the national level.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,7))
plot_change=goglomore.plot(x='date', y='retail_and_recreation_percent_change_from_baseline_reg', ax=ax)
plot_change
plt.xticks( rotation=80)
plt.ylabel('Retail & Recreation Percent Change from Baseline')
plt.title('Evolution of chanes in movements toward retail & recreation since the pandemic outbreak in Turkey',fontdict=None, loc='center')
plt.show()
```
